## While-loop

Schrijf de inhoud van de methode wachtwoord. Deze heeft geen parameters en laat de gebruiker twee wachtwoorden invoeren, waarna hij bekijkt of deze gelijk zijn. Zolang dat niet het geval is wordt er eerst Fout, probeer het opnieuw! uitgeprint en moet de gebruiker het opnieuw proberen.<br>Als de gebruiker het goed doet, komt er Succes! op het scherm te staan.

```python
def wachtwoord():
    while True:
        woord1 = input("Voer het eerste woord in")
        woord2 = input("Voer het tweede woord in")

        if woord1 == woord2:
            print("Succes!")
            break
        else:
            print("Fout, probeer opnieuw.")

wachtwoord()
```

## Objectoriëntatie

Maak een minisysteem voor een kledingwinkel. Maak verschillende klassen voor verschillende soorten kledingstukken, waarbij sommige klassen geërfd worden van andere klassen.
- Maak een klasse Kledingstuk aan, waarin door de constructor een merk, een maat en een prijs geïnitialiseerd worden.
- Maak een klasse Broek en een klasse Shirt, beide erven van de klasse Kledingstuk. Voeg aan Broek een broek_type, en aan Shirt een shirt_type toe (d.m.v. constructorgebruik).
- Maak een klasse Spijkerbroek die erft van broek en een klasse Poloshirt die erft van shirt. Voeg aan Spijkerbroek een blauwtint toe, en aan Poloshirt een kraagsoort (d.m.v. constructorgebruik).
- Implementeer in elke klasse een beschrijvingmethode, waarmee alle informatie over het betreffende kledingstuk wordt afgedrukt.
- Maak van elke klasse een object aan en roep de beschrijvingmethode aan. Controleer nauwkeurig of alles naar behoren werkt.

```python
class Kledingstuk: 
    def __init__(self, merk, maat, prijs):
        self.merk = merk
        self.maat = maat
        self.prijs = prijs

    def beschrijving(self):
        print(f"Merk: {self.merk}, Maat: {self.maat}, Prijs: €{self.prijs}")

class Shirt(Kledingstuk):
    def __init__(self, merk, maat, prijs, shirt_type): 
        super().__init__(merk, maat, prijs)
        self.shirt_type = shirt_type

    def beschrijving(self):
        super().beschrijving()
        print(f"Shirt type: {self.shirt_type}")


class Broek(Kledingstuk):
    def __init__(self, merk, maat, prijs, broek_type):
        super().__init__(merk, maat, prijs)
        self.broek_type = broek_type

    def beschrijving(self):
        super().beschrijving()
        print(f"Broek type: {self.broek_type}")


class Spijkerbroek(Broek):
    def __init__(self, merk, maat, prijs, broek_type, blauwtint):
        super().__init__(merk, maat, prijs, broek_type)
        self.blauwtint = blauwtint

    def beschrijving(self):
        super().beschrijving()
        print(f"Blauwtint: {self.blauwtint}")


class Poloshirt(Shirt):
    def __init__(self, merk, maat, prijs, shirt_type, kraagsoort):
        super().__init__(merk, maat, prijs, shirt_type)
        self.kraagsoort = kraagsoort

    def beschrijving(self):
        super().beschrijving()
        print(f"Kraagsoort: {self.kraagsoort}")


kleding = Kledingstuk("H&M", "M", 29.99)
kleding.beschrijving()

broek = Broek("Levi's", "32", 89.99, "Chino")
broek.beschrijving()

shirt = Shirt("Tommy", "L", 49.99, "T-shirt")  
shirt.beschrijving()

spijker = Spijkerbroek("Levi's", "30", 119.99, "Slim fit", "Donkerblauw")
spijker.beschrijving()

polo = Poloshirt("Lacoste", "M", 79.99, "Polo", "Klassieke kraag") 
polo.beschrijving()
```

Maak nu de tweede kolom ('SECOND_LANGUAGE'). Maak daarvoor een dictionary, met daarin...
- Als keys: alle indexen uit het resultaat van het vorige codeblok.
- Als values: bij de eerste 3 elementen 'FR' (Frankrijk), bij de laatste 2 'DE' (Duitsland).

Maak vervolgens ook hier weer een serie van en voeg ook deze weer horizontaal samen met het rsultaat van de vorige opdracht.<br>
Vergeet niet een passende naam te geven aan de nieuw ontstane kolom.

```python
con = sql.connect("../Data/go_sales.sqlite")
sql_query = "SELECT * FROM sales_staff;"
columns = pd.read_sql(sql_query, con)

sales_staff_code, work_phone, extension, fax, email = [
    columns[col].head(5) for col in ["SALES_STAFF_CODE", "WORK_PHONE", "EXTENSION", "FAX", "EMAIL"]
]
df = pd.concat([sales_staff_code, work_phone, extension, fax, email], axis=1)

firstlang_list = ["EN"] * 5
fll = pd.Series(firstlang_list, name="FIRST_LANGUAGE")

secondlang_values = ["FR"] * 3 + ["DE"] * 2 # ["FR", "FR", "FR"] etc opschrijven is irritant.
secondlang_dict = dict(zip(df.index, secondlang_values)) # wordt {0: 'FR', 1: 'FR', 2: 'FR', 3: 'DE', 4: 'DE'} met zip (doet 2 lists in keys & values doen, en df.index = 5)
sld = pd.Series(secondlang_dict, name="SECOND_LANGUAGE")

df = pd.concat([df, fll, sld], axis=1)

print(df)
con.close() # hoef ik dit nog verder uit te leggen?
```

Maak ten slotte de derde kolom ('THIRD_LANGUAGE') door een dictionary te maken met daarin...
- Als key: de naam van de nieuwe kolom. Zie je het verschil qua keys met de vorige opdracht?
- Als waarde: een lijst met daarin 'NL', 'NL', 'JPN', 'JPN', 'KOR'.

Converteer deze dictionary nu naar een DataFrame en voeg deze horizontaal samen met het resultaat van de vorige opdracht.<br>
Waarom hoef je hierna de nieuw ontstane kolom geen passende naam meer te geven?

```python
con = sql.connect("../Data/go_sales.sqlite")
sql_query = "SELECT * FROM sales_staff;"
sales_staff = pd.read_sql(sql_query, con)

sales_staff_code, work_phone, extension, fax, email = [
    sales_staff[col].head(5) for col in ["SALES_STAFF_CODE", "WORK_PHONE", "EXTENSION", "FAX", "EMAIL"]
]
df = pd.concat([sales_staff_code, work_phone, extension, fax, email], axis=1)

firstlang_list = ["EN"] * 5
fll = pd.Series(firstlang_list, name="FIRST_LANGUAGE")

secondlang_values = ["FR"] * 3 + ["DE"] * 2 # ["FR", "FR", "FR"] etc opschrijven is irritant.
secondlang_dict = dict(zip(df.index, secondlang_values)) # wordt {0: 'FR', 1: 'FR', 2: 'FR', 3: 'DE', 4: 'DE'}. zip maakt tuples van df.index en 1. en omdat het in pairs zijn, kun je dict() gebruiken om van die 5 tuples een dict te maken.
sld = pd.Series(secondlang_dict, name="SECOND_LANGUAGE")

thirdlang_dict = {"THIRD_LANGUAGE": ['NL', 'NL', 'JPN', 'JPN', 'KOR']} # werkt omdat dataframe automatisch de key de naam maakt.
tld = pd.DataFrame(thirdlang_dict) #                                        ^^ hier

df = pd.concat([df, fll, sld, tld], axis=1)

print(df)
con.close() # hoef ik dit nog verder uit te leggen?
```

### Datatypen

Door het attribuut .dtypes van een DataFrame op te vragen krijg je een serie die per kolom het datatype weergeeft. Doe dit bij de originele databasetabel *sales_staff*

```python
con = sql.connect("../Data/go_sales.sqlite")
sql_query = "SELECT * FROM sales_staff;"
sales_staff = pd.read_sql(sql_query, con)

sales_staff_code, work_phone, extension, fax, email, first_name, last_name = [
    sales_staff[col].head(5) for col in ["SALES_STAFF_CODE", "WORK_PHONE", "EXTENSION", "FAX", "EMAIL", "FIRST_NAME", "LAST_NAME"]
]

df = pd.concat([sales_staff_code, work_phone, extension, fax, email, first_name, last_name])

print(df.dtypes)
con.close() # hoef ik dit nog verder uit te leggen?
```

Hier valt op dat elke kolom het datatype 'object' heeft: Python leest dus alles als string. Wiskundige operaties zijn hierdoor niet mogelijk.<br>
We kunnen proberen om kolommen met getallen, bijvoorbeeld de 'extension', te converteren naar een int. Zie onderstaande code.

```python
sales_staff['EXTENSION'] = sales_staff['EXTENSION'].astype(int)
sales_staff.dtypes
```

Dit lukt echter niet, omdat er in de kolom 'EXTENSION' lege waarden zitten die niet geconverteerd kunnen worden naar een int.<br>
Wél kan je deze naar een float converteren, zie onderstaande code:

```python
sales_staff['EXTENSION'] = sales_staff['EXTENSION'].astype(float)
sales_staff.dtypes
```

Als we in de rij van 'EXTENSION' kijken zien we dat de conversie van het datatype nu is gelukt.<br>
Dit is <b>randvoorwaardelijk</b> voor het uitvoeren van wiskundige operaties.<br>
